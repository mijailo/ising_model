```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Gráfica vacía
G = nx.Graph()

#AÑADIR ARISTAS
#Desde una lista
elist = [(1, 2), (2, 3), (1, 4), (4, 2)]
G.add_edges_from(elist)

#AÑADIR NODOS
#Uno por uno:
G.add_node(5)
#Desde una lista
G.add_nodes_from([6, 7])

#DIBUJAR
plt.subplot(121)
nx.draw(G, nodecolor='r', edge_color='b')
```

# Gráficas con una distribución de grados dada

```python
sequence = nx.random_powerlaw_tree_sequence(10, tries=5000)
G = nx.configuration_model(sequence)
plt.subplot()
nx.draw(G)
```

```python
G.edges()
```

```python
H=nx.random_degree_sequence_graph(sequence)
plt.subplot()
nx.draw(H)
```

```python
c = 3  #Conectividad promedio
N = 50 #Número de nodos

s = np.random.poisson(c, N) #Proceso Poisson con suma par
while not sum(s)%2==0:
    s = np.random.poisson(c, N)

G=nx.random_degree_sequence_graph(s,tries=100)
plt.subplot()
nx.draw(G)
```

```python
G = nx.configuration_model(s)
G = nx.Graph(G) #Eliminar aristas paralelas
G.remove_edges_from(nx.selfloop_edges(G))  #Quitar auto-loops
plt.subplot()
nx.draw(G, node_color = 'blue', node_size = 100, width = 0.8, with_labels = False)
```

# Gráfica binomial/Erdös-Rényi/Poisson

```python
n=30 #Número de nodos
p=0.2 #Probabilidad de conexión (=c/N)
H=nx.erdos_renyi_graph(n,p,seed=None,directed=False)
```

```python
%matplotlib notebook

#Grados (lista con nodos y grados correspondientes)
deg=list(H.degree)

nx.draw(H, node_color = 'purple', edge_color='blue',
        node_size=[d[1] * 100 for d in deg],
        width = 0.7, with_labels = False)
```

```python
# Conectividad promedio

np.mean([d[1] for d in deg])
```

```python
# Conectividad promedio (límite)
p*n
```
